```python
# Check what's really in the unemployment table
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()
engine = create_engine(os.getenv('DATABASE_URL'))

df = pd.read_sql("SELECT DISTINCT characteristic, data_type FROM unemployment_monthly", engine)
print(df)
```

```python
# Check how many rows per geography per date
df2 = pd.read_sql("""
    SELECT geography, ref_date, COUNT(*) as row_count, AVG(value) as avg_value
    FROM unemployment_monthly
    WHERE ref_date = '2020-04-01'
    GROUP BY geography, ref_date
    ORDER BY geography
""", engine)
print(df2)
```

```python
df3 = pd.read_sql("""
    SELECT *
    FROM unemployment_monthly
    WHERE ref_date = '2020-04-01'
    AND geography = 'Canada'
    LIMIT 10
""", engine)
print(df3.columns.tolist())
print(df3)
```

```python
# Check what differentiates the 4 rows
df_check = pd.read_csv('data/raw/14100287.csv', usecols=[
    'REF_DATE', 'GEO', 'Labour force characteristics',
    'Gender', 'Age group', 'Data type', 'Statistics', 'VALUE'
])

result = df_check[
    (df_check['REF_DATE'] == '2020-04') &
    (df_check['GEO'] == 'Canada') &
    (df_check['Labour force characteristics'] == 'Unemployment rate') &
    (df_check['Gender'] == 'Total - Gender') &
    (df_check['Age group'] == '15 years and over') &
    (df_check['Data type'] == 'Seasonally adjusted')
]
print(result[['Statistics', 'VALUE']])
```
